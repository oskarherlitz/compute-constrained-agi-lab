# Simulator v0 — Facility Economics

Early exploration of compute/energy constraints.

**Outputs (target):**
- Facility MW
- Annual MWh
- Annual electricity cost

**Inputs:** GPU count, GPU power (kW), utilization, PUE, electricity price ($/kWh)

See `data/sample-scenarios.json` for example inputs and `docs/assumptions.md` for ranges.

```python
import json
from pathlib import Path

def facility_economics(
    gpu_count: int,
    gpu_power_kw: float,
    utilization: float,
    pue: float,
    electricity_price_per_kwh: float,
    annual_hours: int = 8760,
) -> dict:
    """IT load → facility power → annual energy → annual cost."""
    it_load_mw = gpu_count * gpu_power_kw * utilization / 1000
    facility_mw = it_load_mw * pue
    annual_mwh = facility_mw * annual_hours
    annual_cost = annual_mwh * electricity_price_per_kwh
    return {
        "it_load_mw": it_load_mw,
        "facility_mw": facility_mw,
        "annual_mwh": annual_mwh,
        "annual_electricity_cost_usd": annual_cost,
    }

scenarios_path = Path("../data/sample-scenarios.json")
scenarios = json.loads(scenarios_path.read_text())["scenarios"]

for scenario in scenarios:
    result = facility_economics(**scenario["inputs"])
    print(f"\n{scenario['label']}")
    for key, value in result.items():
        print(f"  {key}: {value:,.2f}")
```
